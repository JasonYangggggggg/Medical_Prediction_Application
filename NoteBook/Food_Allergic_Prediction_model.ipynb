```python
df["ASTHMA_START"].fillna(0, inplace = True)
df.ASTHMA_START.value_counts()
```

```python
target = []
for rows in df["ASTHMA_START"]:
    if rows > 0:
        target.append(1)
    else:
        target.append(0)
```

# Asthma prediction base on given condition -> predict base on a person's given information

```python
from sklearn.model_selection import train_test_split
```

```python
df.info()
```

```python
X = df[["BIRTH_YEAR", "GENDER_FACTOR","RACE_FACTOR","ETHNICITY_FACTOR","PAYER_FACTOR","ATOPIC_MARCH_COHORT"]]
X
```

```python
target1 = []
for rows in X["ATOPIC_MARCH_COHORT"]:
    if rows == True:
        target1.append(1)
    else:
        target1.append(0)
```

```python
X["ATOPIC_MARCH_COHORT_1"] = target1
X
```

```python
X.drop('ATOPIC_MARCH_COHORT',
  axis='columns', inplace=True)
X
```

```python
Y = df["target"]
Y.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_features = ["GENDER_FACTOR","RACE_FACTOR","ETHNICITY_FACTOR","PAYER_FACTOR"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,categorical_features)],remainder = "passthrough")
transformer
```

```python
# Assuming df is your DataFrame
shape = X

# Check the dimensions
if len(shape) == 1:
    print("DataFrame is 1D array")
elif len(shape) == 2:
    print("DataFrame is 2D array")
else:
    print("DataFrame has more than 2 dimensions")
```

```python
X_intoTransform = X[["GENDER_FACTOR","RACE_FACTOR","ETHNICITY_FACTOR","PAYER_FACTOR"]]
X_intoTransform.isnull().any()
```
