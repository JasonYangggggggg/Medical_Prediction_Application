```python
import pandas as pd
import numpy as np
import joblib
```

```python
pd.set_option('display.max_columns',None)
```

```python
df_covid = pd.read_csv('Covid Dataset.csv')
df_covid
```

```python
df_covid.info()
```

```python
def transformData(name):
    target = []
    for i in df_covid[name]:
        if i == 'Yes':
            target.append(1)
        elif i == 'No':
            target.append(0)
    df_covid[name] = target
```

```python
def transformGroup(data):
    for (i,j) in data.items():
        transformData(i)
```

```python
transformGroup(df_covid)
```

## Decide what model to use

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X = df_covid.drop("COVID-19",axis=1)
```

```python
y = df_covid["COVID-19"]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model1 = LogisticRegression()
```

```python
model2 = KNeighborsClassifier()
```

```python
model3 = RandomForestClassifier()
```

```python
model1.fit(X_train,y_train)
```

```python
model2.fit(X_train,y_train)
```

```python
model3.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
model2.score(X_test,y_test)
```

```python
model3.score(X_test,y_test)
```

### Random Forest Classifier have the highest score so this model will be used as our prediction model

```python
y_preds = model3.predict(X_test)
```

```python
y_preds
```

```python
print(classification_report(y_test,y_preds))
```
